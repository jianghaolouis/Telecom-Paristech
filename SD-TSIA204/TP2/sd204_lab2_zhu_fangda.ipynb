## Exercise 1 (Linear algebra)

### 1) Check kernel method

```python
def LHS_kernel_trick(X, Y, n, p, labda):
    A_left = np.linalg.solve(X.dot(X.T) + labda * np.identity(n), Y)
    return X.T.dot(A_left)


def RHS_kernel_trick(X, Y, n, p, labda):
    A_right = np.linalg.solve(X.T.dot(X) + labda * np.identity(p), X.T)
    return A_right.dot(Y)
```

#### a) n=100 and p = 2000

```python
n, p, labda = 100, 2000, 10e-5
X = np.random.normal(0, 5, (n, p))
Y = np.random.uniform(-1, 1, n)

U_left = LHS_kernel_trick(X, Y, n, p, labda)
U_right = RHS_kernel_trick(X, Y, n, p, labda)
print("Verify the equility when n = 100, p = 2000:",
      np.allclose(U_left, U_right))
```

#### b) n=2000, p=100

```python
n, p, labda = 2000, 100, 10e-5
X = np.random.normal(0, 5, (n, p))
Y = np.random.uniform(-1, 1, n)
U_left = LHS_kernel_trick(X, Y, n, p, labda)
U_right = RHS_kernel_trick(X, Y, n, p, labda)

print("Verify the equility when n = 2000, p = 100:",
      np.allclose(U_left, U_right))
```

### 2) Compare the time efficient

```python
left_times = []
right_times = []
start, end, step = 100, 2000, 100
n_range = np.arange(start, end, step)
p_range = np.arange(start, end, step)

labda = 10e-5
for n in n_range:
    for p in p_range:
        X = np.random.normal(0, 5, (n, p))
        Y = np.random.uniform(-1, 1, n)

        left_time = time()
        U_left = LHS_kernel_trick(X, Y, n, p, labda)
        left_time = time() - left_time
        left_times.append(left_time)

        right_time = time()
        U_right = RHS_kernel_trick(X, Y, n, p, labda)
        right_time = time() - right_time
        right_times.append(right_time)
```

```python
fig = plt.figure()
size = len(n_range)
z1 = np.reshape(left_times, (size, size))
z2 = np.reshape(right_times, (size, size))
x, y = np.meshgrid(n_range, p_range)

plt.contourf(x, y, z2 - z1, alpha=1)
C = plt.contour(x, y, z2 - z1, alpha=1, cmap=plt.cm.bone)
plt.clabel(C, inline=True, fontsize=10)
```

In this figure, we can find that if n < p , the LHS is more efficient, ohterwise, the RHS_kernel_trickis more efficient. Because for the LHS, we solve the system equation with dimension $n \times n$.For the RHS, we solve the system equation with dimension $p \times p$.  So we can choise the one of two method by the size of n and p in order to get more performence.

## Exercise 3 (Power Method)

### 6) Implementation of power method
The function can be written as the folloing code:

```python
def power_method(X, n, u, v):
    for j in range(n):
        u = np.dot(X, v)
        v = np.dot(np.transpose(X), u)
        v = 1.0 * v / np.linalg.norm(v)
        u = 1.0 * u / np.linalg.norm(u)
    return u, v
```

### 7) Convergence of the power method

In order to illustrate visually the convergence of power method, I calcule the différence between the output of power method and the singluar vectors assocated to the largest singular vectors in each iteration. And then I show it in the figure. 

```python
n, p = 10, 20
u = np.random.uniform(-10, 10, size=n)
v = np.random.uniform(-10, 10, size=p)
X = createMatrixLaPlace(n, p)

u_svd, sigma, v_svd = np.linalg.svd(X)
us_theta = []
vs_theta = []


for n in range(0, 100):
    oldU, oldV = u, v
    u, v = power_method(X, 1, u, v)
    if(n > 2):
        us_theta.append(np.linalg.norm(u - oldU))
        vs_theta.append(np.linalg.norm(v - oldV))


x_ite = np.arange(2, len(us_theta) + 2)
plt.plot(x_ite, us_theta, label="$||u - u_1||$", lw=3)
plt.plot(x_ite, vs_theta, '--',
         label="$||v - v_1||$", lw=3)
plt.legend()
plt.show()
```

From above figure, we can find that the difference between the sigular vectors and outputs of funcitons reduce in function of the numbre of iteration. So we can say that the outputs of power method could approche the singluars vecotrs.
In most case, the output u,vu,v from the algorithm converge to the singular vector associated to the largest singular values of X. Although, in theory, if the initial value of u is perpendicular to the singluar vector, the output will converge to a diffrent vector.

### 9) Largest singluar value
We can use the following formular to get the singular value:
$$
u_1 = \frac{i}{\sigma_1} Xv_i 
$$
So
$$
||\sigma_i u_i|| = ||Av_i||
$$
We can get 
$$
\sigma_i =\frac{ ||Av_i||_2 }{||u_i||_2}
$$

```python
n, p = 20, 60
X = createMatrixLaPlace(n, p)
v = np.random.randn(p)
u = np.random.randn(n)

u, v = power_method(X, 200, u, v)
sing_value = np.linalg.norm(X.dot(v))
u_svd, sigma, v_svd = np.linalg.svd(X)

print("The largest singular value obtained by np.linalg.svd:", sigma[0])
print("The largest singular value obtained by power method:", sing_value)
np.isclose(sing_value, sigma[0])
```

### 10) Seconde largest singular value

We know that 
$$ X = U\Sigma V  = \sum_{i=1}^r \sigma_i u_i v_i^T$$
In order to get the second largest singular value, we can firtly find the largest singular value and its singluar vactor. And then remove the component consisted by largest singluar vector from $X$. And then the second one will become the largest one in the rest of components.It's means the second largest value will become the largest value of the matrix
$$ X - u_1\sigma_1v_i^T $$.

```python
def power_method_first(X, v, n_max):
    """ Singular Value Decomposition.

    This function returns the left singular vector,
    right singluar vector associated to the largest
    singluar value, and its largest singular value
    by power method.

    Args:
        X: A  real matrix to decompose
        u: initial right singular vector
        n_max: max number of iteration
    Returns:
        u_2: the largest left singular vector
        val: the largest value
        v_2: the largest right singular vector
    """
    X, v = np.matrix(X), np.matrix(v).T
    n = 0
    old_v = v * 20
    u = 0
    while(n < n_max and not np.allclose(old_v, v, rtol=1e-08, atol=1e-10)):
        old_v = v
        u = X * v
        v = X.T * u
        v = 1.0 * v / np.linalg.norm(v)
        u = 1.0 * u / np.linalg.norm(u)
        n += 1

    sing_val = np.linalg.norm(X.dot(v))

    return u, sing_val, v


def power_method_second(X, v, n_max):
    """ Singular Value Decomposition.

    This function returns the left singular vector,
    right singluar vector associated to the second largest
    singluar value, and its second largest singular value
    by power method.

    Args:
        X: A  real matrix to decompose
        u: initial right singular vector
        n_max: max number of iteration
    Returns:
        u_2: the second largest left singular vector
        val: the second largest value
        v_2: the second largest right singular vector
    """
    u_1, sigma, v_1 = power_method_first(X, v, n_max)

    X = X - sigma * u_1 * v_1.T

    return power_method_first(X, v, n_max)


def isclose(v1, v2):
    ''' Returns True if two vector are parallel and have equal length.

    Args:
        v1,v2: Input unitary vectors to compare.
    Returns:
        allclose : 
            Returns True if the two vector are parallel 
            within the given tolerance; 
            False otherwise.
    '''

    v1, v2 = np.array(v1).ravel(), np.array(v2).ravel()
    return np.isclose(np.abs(v1.dot(v2)), 1)
```

```python
n, p = 20, 50
X = createMatrixLaPlace(n, p)
U, Sigma, V = np.linalg.svd(X)

v = np.random.randn(p)
u_1, sigma_1, v_1 = power_method_first(X, v, 1000)
u_2, sigma_2, v_2 = power_method_second(X, v, 1000)

print(isclose(u_1, np.matrix(U[:, 0]).T), np.isclose(sigma_1, Sigma[0]))
print(isclose(u_2, U[:, 1]), np.isclose(sigma_2, Sigma[1]))
```
