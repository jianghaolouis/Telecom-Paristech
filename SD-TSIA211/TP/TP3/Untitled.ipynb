## Préparation 2
$$
\begin{array}
\mathcal L_p (x, z^k, r^k) &= f(x) + g(z) + \lambda^T(z-Ax) + \frac{\rho}{2}||z-Ax||_2^2 \\
&= \frac{1}{2}||y-Hx||_2^2 + \beta||z_2||_1 + \mathcal i_{(0,1)^P}(z1) \\
& + \lambda_1^T(z_1-x) + \frac{\rho}{2} ||z_1-x||_2^2 + \lambda_2^T
(z_2-Tx) + \frac{\rho}{2}||z_2 - Tx||_2^2\end{array}
$$
